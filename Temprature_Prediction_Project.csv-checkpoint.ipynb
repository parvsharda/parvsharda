Date is a object datatype.

Hence we will convert this datatype into timestamp.

We will be using pandas to_dateetime function to convert this object to datetime dtype.

.dt.month method will extract the month of the date

# Model Training

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

def train(model,x,y):
    x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state = 42)
    model.fit(x_train,y_train)
    
    #predict the training set
    pred=model.predict(x_test)
    
    #perform cross-validation
    cv_score=cross_val_score(model,x,y,scoring='neg_mean_squared_error',cv=5)
    cv_score=np.abs(np.mean(cv_score))
    
    print("Model Report")
    print("MSE:",mean_squared_error(y_test,pred))
    print("CV Score:",cv_score)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model=LinearRegression(normalize=True)
train(model,x,y)
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
train(model,x,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
train(model,x,y)
```
