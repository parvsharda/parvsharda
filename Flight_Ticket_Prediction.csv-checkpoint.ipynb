# EDA

Date_of_Journey is a object datatype.
Hence we will convert this datatype into timestamp.
We will be using pandas to_dateetime function to convert this object to datetime dtype.

.dt.day method will extract the day of that date
.dt.month method will extract the month of the date

```python
df["Journey_day"]=pd.to_datetime(df.Date_of_Journey, format="%d/%m/%Y").dt.day
```

```python
df["Journey_month"]=pd.to_datetime(df.Date_of_Journey, format="%d/%m/%Y").dt.month
```

```python
df.head()
```

```python
# we can see that now we have two new features in above dataset i.e. Journey_day and Journey_month
# we have data for year 2019 in this dataset and we have already extracted the date and month from Date_of_Journey hence Date_of_Journey is not required.
```

```python
df.drop(columns="Date_of_Journey",inplace=True)
```

```python
#similar to Date_of_journey we can extract values from Dep_Time

#Extracting Hours
df["Dep_hour"]=pd.to_datetime(df["Dep_Time"]).dt.hour

#Extracting Minutes
df["Dep_min"]=pd.to_datetime(df["Dep_Time"]).dt.minute

# we can drop Dep_Time column as we have already extracted the data from it
df.drop(columns="Dep_Time",inplace=True)
```

```python
df.head()
```

```python
#similar to Date_of_journey and Dep_Time we can extract values from Arrival_Time

#Extracting Hours
df["Arrival_hour"]=pd.to_datetime(df["Arrival_Time"]).dt.hour

#Extracting Minutes
df["Arrival_min"]=pd.to_datetime(df["Arrival_Time"]).dt.minute

# we can drop Arrival_Time column as we have already extracted the data from it
df.drop(columns="Arrival_Time",inplace=True)
```

```python
df.head()
```

```python
#Time taken by plane to reach destination is called Duration
# It is the differnce betwwen Departure Time and Arrival time


# Assigning and converting Duration column into list
duration = list(df["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:    # Check if duration contains only hour or mins
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"   # Adds 0 minute
        else:
            duration[i] = "0h " + duration[i]           # Adds 0 hour

duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))   # Extracts only minutes from duration
```

```python
# Adding duration_hours and duration_mins list to train_data dataframe

df["Duration_hours"] = duration_hours
df["Duration_mins"] = duration_mins
```

```python
df.drop(["Duration"], axis = 1, inplace = True)
```

```python
df.head()
```

Handling Categorical Data

One can find many ways to handle categorical data. Some of them categorical data are,

**Nominal data** --> data are not in any order --> **OneHotEncoder** is used in this case
**Ordinal data** --> data are in order --> **LabelEncoder** is used in this case

```python
df["Airline"].value_counts()
```

```python
# As Airline is Nominal Categorical data we will perform OneHotEncoding

Airline = df[["Airline"]]

Airline = pd.get_dummies(Airline, drop_first= True)

Airline.head()
```

```python
df["Source"].value_counts()
```

```python
# As Source is Nominal Categorical data we will perform OneHotEncoding

Source = df[["Source"]]

Source = pd.get_dummies(Source, drop_first= True)

Source.head()
```

```python
df["Destination"].value_counts()
```

```python
# As Destination is Nominal Categorical data we will perform OneHotEncoding

Destination = df[["Destination"]]

Destination = pd.get_dummies(Destination, drop_first = True)

Destination.head()
```

```python
df["Route"]
```

```python
# Additional_Info contains almost 80% no_info
# Route and Total_Stops are related to each other

df.drop(["Route", "Additional_Info"], axis = 1, inplace = True)
```

```python
df["Total_Stops"].value_counts()
```

```python
# As this is case of Ordinal Categorical type we perform LabelEncoder
# Here Values are assigned with corresponding keys

df.replace({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4}, inplace = True)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
# Random search of parameters, using 5 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = reg_rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```
